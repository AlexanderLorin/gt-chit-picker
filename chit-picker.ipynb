```python
#create a list with 'picked' values 
choices = (df['Number'][df['picked']== df['picked'].min()]).to_list()
#pick a random number from the list
lucky = random.choice(choices)
#print the lucky winner's number and name
print( df.loc[df['Number']==lucky,'Name'])
#add 1 to 'picked' column for the winner
df.loc[df['Number']==lucky, 'picked'] +=1
#show the dataframe with udpated value
df
```
